#2020

```python
df_20 = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2020', skiprows = 5)
df_20
```

```python
#Eliminar columna de FECHA
df_2020 = df_20.drop(['FECHA'], axis= 1)
```

```python
df_2020.info()
```

```python
cuantitativas_20 = df_2020.iloc[ : ,[7,8,9,10,11,12,13,17]]
cualitativas_20 = df_2020.iloc[ :, [0,1,2,3,4,5,6,14,15,16]]
```

```python
cuantitativas = cuantitativas_20.fillna(method='bfill')
cualitativas = cualitativas_20.fillna(method='bfill')
```

```python
valores_n = cuantitativas.isnull().sum()
print('Cuantitativas: ',valores_n)
valores = cualitativas.isnull().sum()
print('\nCualitativas: ',valores)
```

```python
fig = plt.figure(figsize = (15,8))
cuantitativas.plot(kind='box', vert=False)
plt.title('Valores atipicos del DataFrame')
plt.show()
```

```python
y = cuantitativas
perc25 = y.quantile(0.25)
perc75 = y.quantile(0.75)
iqr = perc75 - perc25

lim_sup_20 = perc75 + 1.5*iqr
lim_inf_20 = perc25 - 1.5*iqr

print('Limite superior permitido: ', lim_sup_20)
print('\nLimite inferior permitido: ', lim_inf_20)

#Encontrar Outliers
outliers_20 = cuantitativas[(y>lim_sup_20)|(y<lim_inf_20)]
print('\n',outliers_20)
```

```python
#Outliers nulos en el dataframe

cuantitativas2 = cuantitativas[(y<=lim_sup_20)&(y>=lim_inf_20)]
cuantitativas2
```

```python
cuantitativas3 = cuantitativas2.copy()
cuantitativas3 = cuantitativas3.fillna(round(cuantitativas2.mean(),1))
cuantitativas3
```

```python
fecha = df_20['FECHA']
datos_limpios_20 = pd.concat([fecha,cuantitativas3,cualitativas], axis=1)
datos_limpios_20
```

```python
nulos = datos_limpios_20.isnull().sum()
nulos
```
